```julia
using Polynomials
f(x) = 4x^2 + 3*x + sin(2*x)
x0 = -2; x1 = -1; x2 = 1; x3 = 2;
b = [f(x0);f(x1);f(x2);f(x3)]
A = [
    1 x0 x0^2  1;
    1 x1 x1^2 -1;
    1 x2 x2^2  1;
    1 x3 x3^2 -1]
res = A \ b
```

```julia
res_r = map(x -> Float32(x), res)
r = Poly( [res_r[1], res_r[2], res_r[3]])
n = 2
x = linspace(-1, 1, 100); y = f.(x)
plot(x, y, color="red", linewidth=2.0, label="f(x)")
y = r.(x)
plot!(x, y, color="green", label="optymalny $n stopnia")
```
