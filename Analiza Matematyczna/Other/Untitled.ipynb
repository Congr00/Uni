```julia
include("zad 10-8.jl")
using Plots
pyplot()
```

```julia
f(x) = (25x^2 + 1)^(-1)
```

```julia
w1 = Werner_Lagrange(9, f, -1, 1)
@printf("max error: %e\n", max_error(f, w1, -1, 1))

x = linspace(-1,1, 100); y = f.(x)
plot(x, y, color="red", linewidth=2.0, label="f(x)")
y = w1.(x)
plot!(x, y, color="green", label="Lagrange")
```

```julia
w2 = Werner_Lagrange(9, f, -1, 1, ch_nodes(10, -1, 1))
@printf("max error: %e\n", max_error(f, w2, -1, 1))

x = linspace(-1,1, 100); y = f.(x)
plot(x, y, color="red", linewidth=2.0, label="f(x)")
y = w2.(x)
plot!(x, y, color="green", label="Lagrange")
```

```julia
w3 = Werner_Lagrange(9, f, -1, 1, che_nodes(9, -1, 1))
@printf("max error: %e\n", max_error(f, w3, -1, 1))

x = linspace(-1,1, 100); y = f.(x)
plot(x, y, color="red", linewidth=2.0, label="f(x)")
y = w2.(x)
plot!(x, y, color="green", label="Chebyshew")
```
